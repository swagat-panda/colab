## What is Numpy?

### Answer :- Numpy is a linear algebra library for python,which add support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays. 

## Why it is needed in the field of DataScience?
### Answer :- Almost all of the PyData ecosystem rely on numpy as one of their building block.

### It has binding with C libray, so it is fast

## Let's begin our introduction by exploring how to create Numpy arrays.

## Creating Numpy Arrays

### From a Python List

We can create a Numpy array by directly converting a list or list of lists:

```
import numpy as np
```

```
list1 = [10,20,30]
list1
```

```
np.array(list1)
```

```
type(np.array(list1))
```

```
matrix = [[10,20,30],[40,50,60],[70,80,90]]
matrix
```

```
np.array(matrix)
```

## Built-in Methods

### There are lots of built-in ways to generate Arrays in Numpy.
### Lets discuss some of them.


### arange method

this method returns evenly spaced values within a given interval.

```
np.arange(0,10)
```

```
np.arange(0,11,2)
```

```
np.arange(0,11,-2)
```

```
np.arange(0,-11,-2)
```

### zeros and ones methods 

This methods are used to generate arrays of zeros or ones

```
np.zeros(5)
```

```
np.zeros((3,2))
```

```
np.ones(5)
```

```
np.ones((5,5))
```

### linspace method
This method returns evenly spaced numbers over a specified interval.

```
np.linspace(0,10,4)
```

```
np.linspace(0,10,50)
```

### eye method

This method used to create an identity matrix

```
np.eye(5)
```
